```python
def mean_seconds(l):
    return sum(l)/(len(l)*1000)

X = ['Level 1', 'Level 2', 'Level 3', 'Level 4']
completed = [mean_seconds(level_1_success), mean_seconds(level_2_success), mean_seconds(level_3_success), mean_seconds(level_4_success)]
not_completed = [mean_seconds(level_1_fail), mean_seconds(level_2_fail), mean_seconds(level_3_fail), mean_seconds(level_4_fail)]

X_axis = np.arange(len(X))

bars=[]
bars.append(plt.bar(X_axis - 0.2, completed, 0.4, label = 'Completed'))
bars.append(plt.bar(X_axis + 0.2, not_completed, 0.4, label = 'Failed'))


plt.xticks(X_axis,X)
plt.xlabel("Levels")
plt.ylabel("Avg. Time(sec)")
plt.title("Average Time taken for each Level")
plt.legend()
plt.savefig('midterm_analytic_7_time_per_level.png',bbox_inches='tight')
plt.show()
```
